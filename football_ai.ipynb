```python
def bot(goals,red_cards,yellow_cards):
    weight_goals=2.5
    weight_red_cards=1.5
    weight_yellow_cards=0.3
    ans=(goals*weight_goals)+(red_cards*weight_red_cards)+(yellow_cards*weight_yellow_cards)
    return ans
```

```python
ans=bot(2,0,1)
if ans>1:
    print("you proably win")
elif ans<1:
    print("you proably lose")
```

```python
import numpy
def bot2(input):
    weights=numpy.array([2.5,1.5,0.3])
    ans=input.dot(weights)
    return(ans)
    
```

```python
ans=bot2(numpy.array ([2,0,1]))

print(ans)
```
